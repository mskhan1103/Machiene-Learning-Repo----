# Detailed

```python
# 📦 Importing essentials
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 🧠 Load the data
df = pd.read_csv('placement.csv')
x = df['cgpa'].values
y = df['package'].values

# 📊 Compute means
x_mean = np.mean(x)
y_mean = np.mean(y)

# 📐 Calculate slope (m) and intercept (b)
m = np.sum((x - x_mean) * (y - y_mean)) / np.sum((x - x_mean) ** 2)
b = y_mean - m * x_mean

print("Slope (m):", m)
print("Intercept (b):", b)

# 🎯 Make predictions
y_pred = m * x + b

# 📍 Visualize results
plt.figure(figsize=(8, 5))
plt.scatter(x, y, label='Actual', color='blue', alpha=0.6)
plt.plot(x, y_pred, label='Regression Line', color='red')
plt.xlabel('CGPA')
plt.ylabel('Package')
plt.title('Simple Linear Regression - Placement Analysis')
plt.legend()
plt.grid(True)
plt.show()

# ✨ Prediction for a new value (optional)
x_new = float(input("Enter CGPA to predict package: "))
y_new_pred = m * x_new + b
print(f"Your package for {x_new} CGPA is {y_new_pred:.2f}")
```
