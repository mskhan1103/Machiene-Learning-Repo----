```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
# x: independent variable
x = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
# y: dependent variable (quadratic relationship)
y = np.array([3, 6, 15, 28, 45])
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# 1. Generate synthetic curvy data
np.random.seed(0)
x = np.linspace(-5, 5, 30).reshape(-1, 1)              # Shape: (30, 1)
y = 0.5 * x**2 + np.random.randn(30, 1) * 2            # Shape: (30, 1)

# 2. Transform x into polynomial features (degree 2)
poly = PolynomialFeatures(degree=2)
x_poly = poly.fit_transform(x)                         # Shape: (30, 3) [1, x, x^2]

# 3. Fit linear model to polynomial features
model = LinearRegression()
model.fit(x_poly, y)

# 4. Predict using the model
y_pred = model.predict(x_poly)                         # Shape: (30, 1)

# 5. Plot original data and fitted curve
plt.scatter(x, y, color='red', label='Noisy data')
plt.plot(x, y_pred, color='blue', linewidth=2, label='Polynomial Regression')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Polynomial Regression on Curvy Data')
plt.legend()
plt.grid(True)
plt.show()
```
