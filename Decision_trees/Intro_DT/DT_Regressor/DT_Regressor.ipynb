```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor
```

```python
# Example: House Size (sqft) vs Price ($1000)
data = {
    'Size': [500, 750, 1000, 1250, 1500],
    'Price': [100, 150, 200, 250, 300]
}

df = pd.DataFrame(data)

# Features and target
X = df[['Size']]  # 2D array
y = df['Price']   # 1D array
```

```python
model1=DecisionTreeRegressor()
```

```python
model1.fit(X,y)
```

```python
from sklearn.tree import plot_tree
```

```python
plot_tree(model1)
```

```python
new_size = np.array([[1100]])  # Predict for a 1100 sqft house
predicted_price = model1.predict(new_size)
print(f"Predicted price for 1100 sqft: ${predicted_price[0]}k")
```
